```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
fear_greed=pd.read_csv("fear_greed_index.csv")
```

```python
historic= pd.read_csv("historical_data.csv")
```

```python
#drop the timestamp column becaue there we have a date column
fear_greed=fear_greed.drop('timestamp', axis=1)
```

```python
fear_greed
```

```python
fear_greed.info()
```

```python
#normalize date for confirmation
fear_greed['date']=pd.to_datetime(fear_greed['date'], errors='coerce').dt.date
```

```python
fear_greed
```

```python
#confirming classification are in normalize form
fear_greed['classification']=fear_greed['classification'].str.title().str.strip()
```

```python
fear_greed
```

```python
# conforming values are numeric
fear_greed["value"] = pd.to_numeric(fear_greed["value"], errors="coerce")
```

```python
fear_greed
```

```python
#drop null rows
fear_greed=fear_greed.dropna(subset=['value','classification','date'])
fear_greed
```

```python
historic
```

```python
#removing these columns Transaction Hash, Order ID, Trade ID, Timestamp because there no need of these columns in analysis
historic=historic.drop(['Transaction Hash','Order ID', 'Trade ID', 'Timestamp'],axis =1)
```

```python
historic
```

```python
#now we have there timestamp ist column so we need to normalize this column and convert it into date column so we can merge over fear_greed data set with this data set 
historic['Timestamp IST']=pd.to_datetime(historic['Timestamp IST'], format="%d-%m-%Y %H:%M", errors="coerce")
historic
```

```python
#removing time because in fear_greed data set we have only date
historic["date"] = historic["Timestamp IST"].dt.date
historic['date']
```

```python
#just like before convert numeric columns
numeric = ["Execution Price", "Size Tokens", "Size USD", "Closed PnL", "Fee"]
historic[numeric] = historic[numeric].apply(pd.to_numeric, errors="coerce")
```

```python
historic
```

```python
historic=historic.drop('Timestamp IST',axis=1)
```

```python
historic.head()
```

```python
#converting side and direction column into standardize form
historic["Side"] = historic["Side"].str.upper().str.strip()
historic["Direction"] = historic["Direction"].str.title().str.strip()
historic
```

```python
#removing duplicate and null rows
historic = historic.drop_duplicates()
historic = historic.dropna(subset=["Account", "Execution Price", "Size USD", "Closed PnL", "date"])
```

```python
historic
```

```python
# Reset index
historic=historic.reset_index(drop=True)
```

```python
historic.head()
```

```python
#merge the datasets create sentiment 
sentiment = historic.merge(fear_greed[["date", "classification", "value"]], on="date", how="left")
```

```python
sentiment
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Time series: Daily Avg PnL vs. Fear/Greed Index
daily_pnl = sentiment.groupby('date')['Closed PnL'].mean().reset_index()
daily_sentiment = fear_greed[['date', 'value']]
fig, ax1 = plt.subplots(figsize=(12,6))
ax1.plot(daily_pnl['date'], daily_pnl['Closed PnL'], color='blue', label='Avg PnL')
ax1.set_ylabel('Avg PnL (USD)', color='blue')
ax2 = ax1.twinx()
ax2.plot(daily_sentiment['date'], daily_sentiment['value'], color='orange', label='Fear/Greed Value')
ax2.set_ylabel('Fear/Greed Value', color='orange')
plt.title('Daily Avg PnL vs. Fear/Greed Index')
plt.legend()
plt.show()
```

```python
#in plot the bars are representing count of the each sentiment 
plt.figure(figsize=(6,4))
sns.countplot(data=sentiment, x="classification", order=sentiment["classification"].value_counts().index, palette="viridis")
plt.title("Frequency of Sentiment Days")
plt.ylabel("Count")
plt.xlabel("Sentiment")
plt.show()

```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=sentiment, x="classification", hue="Side", palette="Set2")
plt.title("Number of Trades by Sentiment and Side")
plt.ylabel("Trade Count")
plt.xlabel("Sentiment")
plt.show()
```

```python
# above visual represents trade count and sentiments when traders have fear sentiment they buying and selling 
#there trades maxximum and if there have extreme fear they buying and selling trades minimum
sentiment.head()
```

```python
perf_summary = sentiment.groupby('classification')['Closed PnL'].agg(['mean','median','std']).reset_index()
print(perf_summary)
```

```python
plt.figure(figsize=(6,4))
sns.barplot(data=sentiment, x='classification', y='Closed PnL', estimator='mean', ci=None, palette="coolwarm")
plt.title("Average Trader Performance across Sentiments")
plt.ylabel("Average Closed PnL")
plt.show()
```

```python
#above we find mean and make the plot 
plt.figure(figsize=(8,5))
sns.boxplot(data=sentiment, x='classification', y='Closed PnL', palette="Set2")
sns.stripplot(data=sentiment, x='classification', y='Closed PnL', color='blue', size=2, alpha=0.3)
plt.title("Distribution of Trader PnL by Sentiment")
plt.show()

```

```python
#above boxplot and stripplot describes in all the sentiments the profit and loss is neutral
#but there is some outliars which represents some profits and losses are big 
performance_stats = sentiment.groupby("classification")["Closed PnL"].agg( Avg_PnL="mean",Median_PnL="median",Volatility="std",
Count="count").reset_index()
```

```python
print(performance_stats)
```

```python
plt.figure(figsize=(8,5))
plt.bar(performance_stats["classification"],performance_stats["Avg_PnL"],yerr=performance_stats["Volatility"],capsize=5,
color=sns.color_palette("Set2", len(performance_stats))
)
plt.title("Average Trader PnL by Sentiment (with Volatility)")
plt.ylabel("Average Closed PnL")
plt.xlabel("Sentiment")
plt.xticks(rotation=20)
plt.show()
```

```python
#plot the visual on average pnl and in plot there is colored bars shows average pnl of traders it means in all the sentiments average of the traders make profits 
#and the error bars in black shows voletile nature of traders in the greed and extream fear traders makes profit and loss on high level but in the nutral they make less profit and less loss
coin_sentiment_pnl = sentiment.groupby(['Coin','classification'])['Closed PnL'].mean().reset_index()
```

```python
print(coin_sentiment_pnl)
```

```python
pivot_df = sentiment.pivot_table(index='Coin', columns='classification', values='Closed PnL', aggfunc='mean')
plt.figure(figsize=(20,20))
sns.heatmap(pivot_df, annot=True, fmt=".2f", cmap="coolwarm")
plt.title('Average PnL per Coin per Sentiment')
plt.show()

 
```

```python
#above i used many plots but due to coins values there is overlapped data plots always.  now i am doing plot on top10 plots so analysing will be simple
top_coins = sentiment.groupby('Coin')['Closed PnL'].sum().nlargest(10).index
top_df = sentiment[sentiment['Coin'].isin(top_coins)]
```

```python
plt.figure(figsize=(14,7))
sns.stripplot(data=top_df, x='Coin', y='Closed PnL', hue='classification', jitter=True,dodge=True)
plt.title('Top 10 Coins: PnL per Trade with Sentiment')
plt.xlabel('Coin')
plt.ylabel('Closed PnL')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend(title='Sentiment', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# above now this is the plot for top coin based which represents how profit and loss traders makes with coins and on which sentiment they gain profit and loss
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(data=sentiment, x='Coin', y='Closed PnL', hue='classification', split=True)
plt.title('Coin-wise PnL Distribution with Sentiment')
plt.ylabel('Closed PnL')
plt.xlabel('Coin')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
### Key Patterns Uncovered
'''- **PnL by Sentiment**: Highest avg PnL in Extreme Greed (67.89) with moderate volatility; Fear has high trade volume (61,837) but volatile PnL (std 935).
- **Trade Behavior**: More Buy/Sell in Fear, indicating panic or opportunity hunting.
- **Coin-Specific**: Top coins (e.g., FARTCOIN) lose in Fear but gain in Greed; outliers show high-reward potential.
- **Volatility**: High in Greed (1116), low in Neutral (517)—risk-reward varies.'''


### Smarter Trading Strategies
'''1. **Greed/Extreme Greed**: Boost long positions (Buy) for profits; use stop-loss for volatility.
2. **Fear/Extreme Fear**: Hedge or short-sell; reduce volume to avoid losses.
3. **Neutral**: Steady trading on stable coins for low-risk returns.
4. **General**: Diversify based on coin patterns; monitor fees in high-volume sentiments.
5. **Advanced**: Use sentiment thresholds (value >70) for automated alerts; backtest on historic data.'''
```
